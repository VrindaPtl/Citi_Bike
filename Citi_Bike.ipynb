```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# read csv file
citibike_df1 = pd.read_csv('/Users/vrindapatel/Desktop/New York Citi Bike/JC-202402-citibike-tripdata.csv')
citibike_df2 = pd.read_csv('/Users/vrindapatel/Desktop/New York Citi Bike/JC-202403-citibike-tripdata.csv')
citibike_df3 = pd.read_csv('/Users/vrindapatel/Desktop/New York Citi Bike/JC-202404-citibike-tripdata.csv')

# create a DataFrame from the csv files
citibike_df = pd.concat([citibike_df1, citibike_df2, citibike_df3])
citibike_df.head()
```

```python
# clean up the data
citibike_df = citibike_df.dropna()
citibike_df = citibike_df.drop_duplicates()
citibike_df.head()
```

```python
# create a new column "Distance" to calculate the distance between start and end stations
citibike_df['Distance'] = np.sqrt((citibike_df['start_lat'] - citibike_df['end_lat'])**2 + (citibike_df['start_lng'] - citibike_df['end_lng'])**2)
citibike_df.head()
```

```python
import calendar
import pandas as pd

# Convert 'started_at' to datetime if not already
citibike_df['started_at'] = pd.to_datetime(citibike_df['started_at'])

# Create a "Month" column to store the name of the month of the trip
citibike_df['Month'] = citibike_df['started_at'].dt.month.apply(lambda x: calendar.month_name[x])

# Create a "Day" column to store the name of the day of the week of the trip
citibike_df['Day'] = citibike_df['started_at'].dt.day_name()

# Display the first few rows to verify
citibike_df.head()
```

```python
# Ensure 'started_at' and 'ended_at' are in datetime format
citibike_df['started_at'] = pd.to_datetime(citibike_df['started_at'])
citibike_df['ended_at'] = pd.to_datetime(citibike_df['ended_at'])

# Now split the 'started_at' and 'ended_at' columns into 'Date' and 'Time' columns
citibike_df['Date'] = citibike_df['started_at'].dt.date
citibike_df['Time'] = citibike_df['started_at'].dt.time
citibike_df['End Date'] = citibike_df['ended_at'].dt.date
citibike_df['End Time'] = citibike_df['ended_at'].dt.time

# Display the first few rows to verify
citibike_df.head()
```

```python
# save the cleaned data to a new csv file

citibike_df.to_csv('/Users/vrindapatel/Desktop/New York Citi Bike/citibike-tripdata.csv', index=False)
```
